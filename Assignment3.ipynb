```python
def monthIndexing(index):
    year = 1749 + index // 12
    month = (index % 12) + 1 
    return month, year 
```

```python
def monthIndex(month, year):
    index = (year - 1749) * 12 + (month - 1)
    
    if year < 1749 or year > 1983 or month < 1 or month > 12: #is the date valid??? u are valid girl 
        return print (f"ERROR: Invalid Date Input. This dataset has {len(sunspots)} elements, corresponding to months between {monthIndexing(0)} and {monthIndexing(len(sunspots)-1)}.")
    
    else: 
        return index
```

```python
def SunspotsYearly(year):
    start = monthIndex(1, year)
    end = monthIndex(12, year)
    
    if end is None or start is None:
        return "Date range provided is invalid."
    
    return sunspots[start:end+1]
```

```python
def get_average(year):
    s = monthIndex(1, year)
    e = monthIndex(12, year)
    n = 0  

    if s is None or e is None: 
        return "Date provided is invalid."
        
    for i in range(s, e + 1): 
        n += sunspots[i] 
        
    average = n / 12 
    return average
```

```python
def MinMaxDiff(year1, year2):
    sunspots_year1 = SunspotsYearly(year1)
    sunspots_year2 = SunspotsYearly(year2)
    
    if len(sunspots_year1) != 12 or len(sunspots_year2) != 12:
        return "Error: One of these years doesn't have 12 months of data."
    
    min_diff = float('inf')
    max_diff = float('-inf')
    min_month = max_month = None
    
    for i in range(12):
        diff = abs(sunspots_year1[i] - sunspots_year2[i])
        
        if diff < min_diff:
            min_diff = diff
            min_month = i + 1 
        
        if diff > max_diff:
            max_diff = diff
            max_month = i + 1
    
   
    print (f"Smallest difference: {min_diff} in month {min_month}")
    print (f"Largest difference: {max_diff} in month {max_month}")
```

```python
def get_year_and_month(count):
    indices = np.where(sunspots == count)[0]
    
    if len(indices) == 0:
        print(f"No sunspot counts found for the value {count}.")
        return

    print (f'Total observations of {count} in dataset: {len(indices)}')
    for i in indices:
        month, year = monthIndexing(i)
        print(f"Sunspot count of {count} observed in {month}/{year} (Index: {i})")
```

```python
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

for year in range(1900, 1911):  
    SunspotCounts = SunspotsYearly(year)
    
    MaxCount = np.max(SunspotCounts)
    MaxMonthIndex = np.argmax(SunspotCounts)
    
    # Convert month index to string format
    MaxMonthStr = months[MaxMonthIndex]
    
    # Print the result
    print(f"Year: {year}, Month: {MaxMonthStr}, Sunspot Count: {MaxCount}")
```

```python
x_labels = []
for i in range(0, len(sunspots), 240): 
    month, year = monthIndexing(i)
    x_labels.append(f'{year}')


plt.figure(figsize=(12, 6))
plt.plot(sunspots, linestyle='--', color='blue') 
plt.title("Sunspot Counts from 1749 to 1983")
plt.xlabel("Time (Year)")
plt.ylabel("Number of Sunspots")
plt.xticks(ticks=np.arange(0, len(sunspots), 240), labels=x_labels,)
plt.show()
```

```python
start_1800 = monthIndex(1, 1800)
end_1850 = monthIndex(12, 1850)
sunspots_1800_1850 = sunspots[start_1800:end_1850 + 1]


start_1900 = monthIndex(1, 1900)
end_1950 = monthIndex(12, 1950)
sunspots_1900_1950 = sunspots[start_1900:end_1950 + 1]

x_labels = []
for i in range(0, len(sunspots_1800_1850), 12):  
    year_1800 = 1800 + i // 12
    year_1900 = 1900 + i // 12
    x_labels.append(f'{year_1800}-{year_1900}')

plt.figure(figsize=(14, 7))
plt.plot(sunspots_1800_1850, label='1800-1850', linestyle='-', color='green')
plt.plot(sunspots_1900_1950, label='1900-1950', linestyle='--', color='red')
plt.xticks(ticks=np.arange(0, len(sunspots_1800_1850), 12), labels=x_labels, rotation=90)
plt.title("Sunspot Counts: 1800-1850 vs. 1900-1950")
plt.xlabel("Year (1800-1850) and (1900-1950)")
plt.ylabel("Number of Sunspots")
plt.legend()

plt.savefig("sunspot_counts_1800_1850_vs_1900_1950.png")
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(14, 10))

x_labels_1800_1850 = [f'{1800 + i // 12}' for i in range(0, len(sunspots_1800_1850), 12)]

x_labels_1900_1950 = [f'{1900 + i // 12}' for i in range(0, len(sunspots_1900_1950), 12)]

ax1.plot(sunspots_1800_1850, linestyle='-', color='green')
ax1.set_title("Sunspot Counts from 1800 to 1850")
ax1.set_xlabel("Year (1800-1850)")
ax1.set_ylabel("Number of Sunspots")
ax1.set_xticks(np.arange(0, len(sunspots_1800_1850), 12))
ax1.set_xticklabels(x_labels_1800_1850, rotation=45)

ax2.plot(sunspots_1900_1950, linestyle='--', color='red')
ax2.set_title("Sunspot Counts from 1900 to 1950")
ax2.set_xlabel("Year (1900-1950)")
ax2.set_ylabel("Number of Sunspots")
ax2.set_xticks(np.arange(0, len(sunspots_1900_1950), 12))
ax2.set_xticklabels(x_labels_1900_1950, rotation=45)

plt.tight_layout()

plt.savefig("sunspot_counts_1800_1850_and_1900_1950_subplots.png")
plt.show()
```

```python
start_1801 = monthIndex(1, 1801)
end_1900 = monthIndex(12, 1900)
sunspots_19s = sunspots[start_1801:end_1900 + 1]

sunspots_19s_shifted = np.roll(sunspots_19s, shift=12)
sunspots_19s_shifted[:12] = 0 

years = np.arange(1801, 1901, 10)
x_labels = [str(year) for year in years]
x_ticks = [(year - 1801) * 12 for year in years]


plt.figure(figsize=(14, 7))
plt.plot(sunspots_19s, linestyle='-', color='blue', label='Original Data')
plt.plot(sunspots_19s_shifted, linestyle='--', color='orange', label='12-Month Lagged Data')

plt.xticks(ticks=x_ticks, labels=x_labels, rotation=45)
plt.title("Sunspot Counts with 12-Month Lag Comparison (1801-1900)")
plt.xlabel("Year")
plt.ylabel("Number of Sunspots")
plt.legend()

plt.show()
```
